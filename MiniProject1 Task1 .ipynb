```python
import sklearn 
from sklearn.datasets import load_files
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.model_selection import train_test_split  # Question 5
from sklearn.naive_bayes import MultinomialNB # Question 6

from sklearn import metrics # Question 7 this imports confusion matrix as well so no Need for it below
from sklearn.metrics import confusion_matrix
from sklearn.metrics import classification_report #7c
from sklearn.metrics import accuracy_score
from sklearn.metrics import f1_score



# Question 3 Loading the corpus
new_corpus = load_files('C:\\Users\\Gaith\\Downloads\\BBC_dataset\\BBC', load_content=True, encoding='latin1') # Loading the files
#print(new_corpus.filenames)     print(new_corpus.DESCR)        print(new_corpus.data)


# Question 5 Data Split -> train_test_split: Split arrays or matrices into random train and test subsets
# Splitting the dataset into 80% for training and 20% for testing
X_train, X_test, Y_train, Y_test = train_test_split(new_corpus.data, new_corpus.target, test_size=0.2, random_state=None)

print('---------------------------------------------------------------------')


# Question 4
vectorizer = CountVectorizer()
vectorizer.fit(X_train)
#print(vectorizer.vocabulary_) # prints Count of each word
matrix = vectorizer.transform(X_train)  # Transforming training data
#print(matrix.shape)   # Prints (1780, 26993)
#print(matrix.toarray())

X_test_docmatrix = vectorizer.transform(X_test)    # Might delete

# Question 6
AI = MultinomialNB() # Instantiating a Multinomial Bayes
AI.fit(matrix, Y_train)
MultinomialNB()

Y_prediction_class = AI.predict(X_test_docmatrix) # Class Prediction



# Question 7
 

# Question 7 (b)
ConfusionMatrix = metrics.confusion_matrix(Y_test, Y_prediction_class)
print('Question 7(b)','Confusion Matrix: ')
print(ConfusionMatrix, '\n')


# Question 7 (c)
target_names = ['business', 'entertainment', 'politics', 'sport', 'tech']
print('\n', 'Question 7 (c) Classification Report')
print(classification_report(Y_test, Y_prediction_class, target_names=target_names))
#f.write(classification_report(Y_test, Y_prediction_class, target_names=target_names))


# Question 7 (d)
Accuracy = metrics.accuracy_score(Y_test, Y_prediction_class)
MacroF1score = f1_score(Y_test, Y_prediction_class, average='macro')
WeightedF1score = f1_score(Y_test, Y_prediction_class, average='weighted')

print('\n', 'Question 7 (d)')
print('Accuracy = ', Accuracy)
print('MacroF1score = ', MacroF1score)
print('WeightedF1score = ', WeightedF1score)

# Question 7 (e)
#priorProb = AI.class_prior()
classParameters = AI.get_params()

print('\n', 'Question 7 (e)')
#print('priorProb', priorProb)

print('Parameters: ', classParameters)

print('---------------------------------------------------------------------', '\n')



# Question 9
Multinom9 = MultinomialNB(alpha = 0.0001) # Instantiating a Multinomial Bayes
Multinom9.fit(matrix, Y_train)
MultinomialNB()

Y_pred_class9 = Multinom9.predict(X_test_docmatrix) # Class Prediction

# Question 7 (b)
ConfusionMatrix9 = metrics.confusion_matrix(Y_test, Y_pred_class9)
print('Question 7(b) from (Question 9)','Confusion Matrix: ')
print(ConfusionMatrix9, '\n')


# Question 7 (c)
target_names9 = ['business', 'entertainment', 'politics', 'sport', 'tech']
print('\n', 'Question 7 (c) from (Question 9) Classification Report')
print(classification_report(Y_test, Y_pred_class9, target_names=target_names9))


# Question 7 (d)
Accuracy9 = metrics.accuracy_score(Y_test, Y_pred_class9)
MacroF1score9 = f1_score(Y_test, Y_pred_class9, average='macro')
WeightedF1score9 = f1_score(Y_test, Y_pred_class9, average='weighted')

print('\n', 'Question 7 (d) from (Question 9)')
print('Accuracy = ', Accuracy9)
print('MacroF1score = ', MacroF1score9)
print('WeightedF1score = ', WeightedF1score9)

# Question 7 (e)
#Multinom9.class_prior()
classParameters9 = Multinom9.get_params()

print('\n', 'Question 7 (e) from (Question 9)')
#print('priorProb', priorProbS)

print('Parameters: ', classParameters9)
print('---------------------------------------------------------------------', '\n')




# Question 10
Multinom10 = MultinomialNB(alpha = 0.9) # Instantiating a Multinomial Bayes
Multinom10.fit(matrix, Y_train)
MultinomialNB()

Y_pred_class10 = Multinom10.predict(X_test_docmatrix) # Class Prediction

# Question 7 (b)
ConfusionMatrix10 = metrics.confusion_matrix(Y_test, Y_pred_class10)
print('Question 7(b) from (Question 10)','Confusion Matrix: ')
print(ConfusionMatrix10, '\n')


# Question 7 (c)
target_names10 = ['business', 'entertainment', 'politics', 'sport', 'tech']
print('\n', 'Question 7 (c) from (Question 10) Classification Report')
print(classification_report(Y_test, Y_pred_class10, target_names=target_names10))
#f.write(classification_report(Y_test, Y_pred_class, target_names=target_names))


# Question 7 (d)
Accuracy10 = metrics.accuracy_score(Y_test, Y_pred_class10)
MacroF1score10 = f1_score(Y_test, Y_pred_class10, average='macro')
WeightedF1score10 = f1_score(Y_test, Y_pred_class10, average='weighted')

print('\n', 'Question 7 (d) from (Question 10)')
print('Accuracy = ', Accuracy10)
print('MacroF1score = ', MacroF1score10)
print('WeightedF1score = ', WeightedF1score10)

# Question 7 (e)
#priorProb10 = Multinom10.class_prior()
classParameters10 = Multinom10.get_params()

print('\n', 'Question 7 (e) from (Question 10)')
#print('priorProb', priorProb10)
print('Parameters: ', classParameters10)



```
