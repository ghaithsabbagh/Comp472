```python
import matplotlib.pyplot as plot
import pandas as pd
import csv
```

```python
analysis = pd.read_csv('C:\A\Classes\Comp472\\analysis.csv')  # reading the synonyms csv file as a Dataframe
#analysis.head(10)
```

```python
accuracy = analysis["0.886075949"]

glove_twitter_50 = accuracy[0]
glove_wiki_gigaword_50 = accuracy[1]
glove_twitter_100 = accuracy[2]
glove_twitter_200 = accuracy[3]

```

```python
x_axis = ["word2vec-google-news-300", "glove-twitter-50", "glove-wiki-gigaword-50", "glove-twitter-100", "glove-twitter-200"] 
y_axis = [0.886075949, glove_twitter_50, glove_wiki_gigaword_50, glove_twitter_100, glove_twitter_200]

ax= plot.subplot()
plot.bar(x_axis, y_axis, width=0.8, align='center')
plot.xlabel('Model')
plot.ylabel('Performance (Accuracy)')
plot.title('Comparison of performance of models')
plot.tight_layout()
plot.setp(ax.get_xticklabels(), rotation=30, ha='right')

plot.savefig("Model Performance.pdf")
```
