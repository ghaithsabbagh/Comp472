```python
synonyms = pd.read_csv('C:\A\Classes\Comp472\\synonyms.csv')  # reading the synonyms csv file as a Dataframe
synonyms.head(10)
```

```python
question_words = synonyms["question"].values  
question_words   # List to hold the question words

answer_words = synonyms["answer"].values
answer_words  # List to hold the answers

guessword_1 = synonyms["0"].values
guessword_1  # List to hold the first column of answers

guessword_2 = synonyms["1"].values
guessword_2

guessword_3 = synonyms["2"].values
guessword_3

guessword_4 = synonyms["3"].values
guessword_4
```

```python
def simFunc(index):
    if question_words[index] in model:
        if guessword_1[index] in model and guessword_2[index] in model and guessword_3[index] in model and guessword_4[index] in model:
            v1 = model.similarity(question_words[index], guessword_1[index])
            v2 = model.similarity(question_words[index], guessword_2[index])
            v3 = model.similarity(question_words[index], guessword_3[index])
            v4 = model.similarity(question_words[index], guessword_4[index])
            arr = [v1, v2, v3, v4]
            return(np.argmax(arr))   #return the index of the largest similraaity index        
        else:
            return -1
    else:
        return -1
        
#Creating a 2D list to hold the result           
rows, cols = (80, 4)
myMatrix = []#2D list

for i in range(rows):
    exists = True   # Boolean to check if a word exists in model
    col = []
    for j in range(cols):
        if j == 0:
            col.append(question_words[i])
        elif j == 1:
            col.append(answer_words[i])
        elif j == 2:
            new_index = simFunc(i)
            if  new_index == -1:
                exists = False
                col.append(guessword_4[i])
            else:
                if new_index == 0:
                    col.append(guessword_1[i])
                elif new_index == 1:
                    col.append(guessword_2[i])
                elif new_index == 2:
                    col.append(guessword_3[i])
                elif new_index == 3:
                    col.append(guessword_4[i])
        else:
            if exists == False:
                col.append('guess')
            else:
                if answer_words[i] == col[j-1]:
                    col.append('correct')
                elif answer_words[i] != col[j-1]:
                    col.append('wrong')
            
    myMatrix.append(col)
#print(myMatrix)
myMatrix_np = np.array(myMatrix)          #convert 2D list into numpy 2D array
print(myMatrix_np)
myMatrix_df = pd.DataFrame(myMatrix_np)  #converting numpy array into a pandas dataframe
myMatrix_df.to_csv (r'C:\A\Classes\Comp472\Corpuses\csv\glove-twitter-100-details.csv', index = None, header=True)
```

```python
import csv

vocab_size = len(model) # size of vocab

df = pd.read_csv(r'C:\A\Classes\Comp472\Corpuses\csv\glove-twitter-100-details.csv')
label_df = df['3'].values  #isolate the last column which contains correct/guess
label_df
C=0
for i in range(len(label_df)):
    if label_df[i] == 'correct':
        C += 1
print(C)  #the number of correct labels

count_guess=0
for i in range(len(label_df)):
    if label_df[i] == 'guess':
        count_guess += 1
print(count_guess)                       #the number of guess labels  (1)
V = len(label_df) - count_guess          #the number of questions that model answered without guessing (79)
print(V)
accuracy = C/V
print(accuracy)

# open the file in the write mode
f = open(r'C:\A\Classes\Comp472\analysis.csv', 'a')

# create the csv writer
writer = csv.writer(f)

# write a row to the csv file
writer.writerow(["glove-twitter-100", vocab_size, C, V, accuracy])

# close the file
f.close()
```
